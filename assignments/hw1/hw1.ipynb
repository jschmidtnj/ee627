```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data.csv')
print(df.columns)
unused_cols = ['Date']
df = df.drop(columns=unused_cols)
df
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
factors = ['Mkt-RF', 'SMB', 'HML', 'RF', 'Mom']

max_factors, min_factors = [], []
used_cols = []
for col in df.columns:
    if col in factors:
        continue
    used_cols.append(col)
    max_factor, min_factor = None, None
    max_factor_val, min_factor_val = None, None
    for factor in factors:
        curr_val = correlation_matrix[col][factor]
        if max_factor is None or curr_val > max_factor_val:
            max_factor_val = curr_val
            max_factor = factor
        if min_factor is None or curr_val < min_factor_val:
            min_factor_val = curr_val
            min_factor = factor
    max_factors.append(max_factor)
    min_factors.append(min_factor)

factors_df = pd.DataFrame([max_factors, min_factors], columns=used_cols, index=['positive', 'negative'])
print('max & min correlation factors:', factors_df.to_numpy())

rf_data = correlation_matrix['RF'][[col for col in df.columns if col not in factors]]
# print(rf_data)
avg_rf_correlation = np.mean(rf_data)
print('average rf correlation:', avg_rf_correlation)

factors_df
```

## part 2

Suppose that the daily value of a financial time series follows the model 

$X_t=0.01 + 0.2 X_{t-2} + a_t$

where $\{ a_t \}$ is a Gaussian white noise series with mean zero and variance 0.02.

- What are the mean and variance of the return series $X_t$?

### mean
  
$a_t$ non-deterministic, the rest deterministic  
  
$E\{X_t\} = E\{0.01 + 0.2 X_{t-2} + a_t\}$  
= $0.01 + E\{0.2 X_{t-2}\} + E\{a_t\}$  
= $0.01 + 0.2 E\{X_{t-2}\} + E\{a_t\}$  
$E\{a_t\} = 0$  
stationary => $E\{X_t\} = E\{X_{t-2}\}$  
$E\{a_t\} = 0.01 + 0.2 \cdot E\{a_t\} + 0$  
  
mean = E\{a_t\} = <mark style="background-color: lightblue">0.0125</mark>  
  
### variance

$\mu = E\{X_t\}$  
$X_t = 0.01 + 0.2 X_{t-2} + a_t$  
$(X_t - \mu) = 0.2 (X_{t - 2} - \mu) + a_t$  

$E\{X_t - \mu\}^2 = (0.2 (X_{t - 2} - \mu) + a_t)^2$  
$= E\{0.04 \cdot (X_{t - 2} - \mu)^2\} + E\{0.4 a_t \cdot (X_{t - 2} - \mu)\} + E\{a_t^2\}$  
$a_t$ is noise for today, independent from $X_{t-2}$. correlation is 0, second term = 0  
because it's stationary, $X_{t - 2} - \mu = variance$  
$var = 0.2^2 \cdot var + \sigma_a^2$, $\sigma_a^2$ is the noise level ($E\{a_t^2\}$)  

variance = <mark style="background-color: lightblue">$\frac{\sigma_a^2}{0.96}$</mark>  

- Compute the lag-1 and lag-2 autocorrelations of $X_t$. 

$X_t = 0.01 + 0.2 X_{t-2} + a_t$  
remove mean ($X_t = X_t - \mu$)  
$X_t - \mu = 0.01 + 0.2 X_{t-2} + a_t - \mu$  
$X_t - \mu = 0.2(X_{t-2} - \mu) + a_t$  
$\overline{X_t} = 0.2\overline{X}_{t-2} + a_t$  
  
auto-correlation:  
$E\{\overline{X_t} \cdot \overline{X}_{t-l}\} = E\{(0.2\overline{X}_{t-2} + a_t) \overline{X}_{t-l}\}$  
$E\{\overline{X_t} \cdot \overline{X}_{t-l}\} = E\{0.2 \cdot \overline{X}_{t-2} \cdot \overline{X}_{t-l}\}$  

lag 1:  
$E\{\overline{X_t} \cdot \overline{X}_{t-1}\} = E\{0.2 \cdot \overline{X}_{t-2} \cdot \overline{X}_{t-1}\}$  
lag 2:  
$E\{\overline{X_t} \cdot \overline{X}_{t-2}\} = E\{0.2 \cdot \overline{X}_{t-2} \cdot \overline{X}_{t-2}\}$
